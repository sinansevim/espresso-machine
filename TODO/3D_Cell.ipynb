```python
from src import espresso_machine as esma
```

```python
#Step 1 - Initialize model
model = esma.project(project_id="P") #Define project
model.set_cores(8) #Define number of prcessing cores
model.get_structure(format='poscar',path='./Research/BP',name='P.poscar') #Load structure
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cell = np.array(model.config['pw']['cell_parameters']).astype(float)
```

```python
%matplotlib widget
```

```python
ax = plt.axes(projection='3d')
atoms = np.array(model.config['pw']['atomic_positions']).T[1:].T.astype(float)
cartesian = np.dot(atoms,cell)
for i in cell:
    ax.plot([0,i[0]],[0,i[1]],[0,i[2]])
for i in cartesian:
    ax.scatter(i[0],i[1], i[2],s=100,c='red')
plt.axis("equal")
plt.show()
```
